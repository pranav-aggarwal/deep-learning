```python
import pandas as pd
import re

#df = pd.read_csv(r'access_log_Jul95.txt', sep='\ (?!-)(?!\s*\S+")(?!\s+)', engine="python")
#df = pd.read_csv(r'access_log_Jul95.txt', sep='\ (?=\[)|(?<="\w)\s|\s(?=\/)|(?<=\])\ |(?<=")\ |(?<=\d)\ (?=\d)', engine="python")


df = pd.read_csv(r'access_log_Jul95.txt', sep='\ (?=\[)|(?<="\w\w\w)\s(?=\/)|(?<="\w\w\w\w)\s(?=\/)|(?<="\w\w\w\w\w)\s(?=\/)|(?<=\])\ |(?<=")\ |(?<=\d)\ (?=\d)', engine="python", header=None)
```

```python
df.columns = ['host', 'time', 'request_type', 'url', 'response', 'size']
```

```python
df
```

```python
df['request_type'] = df['request_type'].str.replace(r'"', '', regex=True)
```

```python
df['url'] = df['url'].str.replace(r'"', '', regex=True)
df['host'] = df['host'].str.replace(r'(?<=\S)\ +.*', '', regex=True)
df['time'] = df['time'].str.replace(r'[\[\]]', '', regex=True)
df['time_r'] = pd.to_datetime(df['time'], format='%d/%b/%Y:%H:%M:%S %z', utc=True)
```
